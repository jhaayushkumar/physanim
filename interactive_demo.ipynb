# PhysAnim Interactive Demo
Run this notebook to try out real-time physics interactions.

```python
%matplotlib widget
import physanim
from physanim.systems import DoublePendulum, SpringMass
from physanim import interact
```

## Interactive Double Pendulum
Adjust the lengths and masses to see how the chaos changes.

```python
interact(DoublePendulum, 
         L1=(0.5, 3.0, 0.1), 
         L2=(0.5, 3.0, 0.1), 
         m1=(0.1, 5.0, 0.1), 
         m2=(0.1, 5.0, 0.1))
```

## Interactive Spring Mass
Tune the spring constant (k) and damping to design a shock absorber.

```python
interact(SpringMass, 
         k=(1.0, 50.0), 
         damping=(0.0, 5.0), 
         m=(0.5, 10.0))
```
